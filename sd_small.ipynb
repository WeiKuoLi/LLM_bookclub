```python
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("OFA-Sys/small-stable-diffusion-v0", torch_dtype=torch.float16).to("cuda")

#model_id = "OFA-Sys/small-stable-diffusion-v0/"
#pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
#pipe = pipe.to("cuda")

prompt = "an apple, 4k"
image = pipeline(prompt, num_inference_steps=15).images[0]

image.save("apple.png")
```

```python
pipeline.save_pretrained("sd_small")
```
